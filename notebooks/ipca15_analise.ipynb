```python
import basedosdados as bd
import numpy as np
import pandas as pd
import plotly.express as px

import matplotlib.pyplot as plt
plt.rcParams['svg.fonttype'] = 'none'

import seaborn as sns
pd.options.display.max_columns = 999
pd.options.display.max_rows = 1999
```

```python
query = """
SELECT * FROM `basedosdados.br_ibge_inflacao.ipca15`
"""
df = bd.read_sql(query, billing_project_id='basedosdados')
```

```python
df['ano_mes'] = df['ano'].astype(str) + '-' + df['mes'].astype(str)
```

```python
df.head()
```

```python
fig = px.line(df, x='ano_mes', 
              y=['indice',
                 'variacao_mes',
                 'variacao_tres_meses',
                 'variacao_semestral',
                 'variacao_anual',
                 'variacao_doze_meses',
                 
            ])
fig = fig.update_traces(line=dict(width=3))
fig = fig.update_layout(
    hovermode = 'x unified',
    height=720,
#         width=1080,
)

fig.show()
```
