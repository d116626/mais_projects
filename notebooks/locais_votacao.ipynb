```python
def download_data():
    ## DOWNLOAD DATA
    urls = [
        'http://pindograma-dados.s3.amazonaws.com/mapa/polling-stations/local-votacao-2008.csv',
        'http://pindograma-dados.s3.amazonaws.com/mapa/polling-stations/local-votacao-2010.csv',
        'http://pindograma-dados.s3.amazonaws.com/mapa/polling-stations/local-votacao-2012.xlsx',
        'http://pindograma-dados.s3.amazonaws.com/mapa/polling-stations/local-votacao-2014.csv',
        'http://pindograma-dados.s3.amazonaws.com/mapa/polling-stations/local-votacao-2016.csv',
        'http://pindograma-dados.s3.amazonaws.com/mapa/polling-stations/local-votacao-08-08-2018.csv',
        'http://pindograma-dados.s3.amazonaws.com/mapa/polling-stations/local-votacao-2020-nov.csv'
    ]


    for url in urls:

        ## download from URL
        save_path = '../data/locais_votacao/input/'    
        wget.download(url, out=save_path)

        ## rename files
        ano = re.findall(r'\d\d\d\d', url)[0]
        old_name = url.split('/')[-1]
        extension = old_name.split('.')[-1]
        new_name = f'local_votacao_{ano}.{extension}'
        os.rename(save_path + old_name, save_path + new_name)
        
        
def convert_2012():
    df = pd.read_excel('../data/locais_votacao/input/local_votacao_2012.xlsx', dtype=str, sheet_name='Plan2')
    df.to_csv('../data/locais_votacao/input/local_votacao_2012.csv',sep=';', index=False, encoding='latin1')
    os.remove('../data/locais_votacao/input/local_votacao_2012.xlsx')

# download_data()
# convert_2012()
```
